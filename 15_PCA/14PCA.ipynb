#  2. Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target variable
X = df.drop('Alcohol', axis=1)
y = df['Alcohol']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# Scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', label='Individual Variance')
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='s', label='Cumulative Variance', color='orange')
plt.axhline(y=0.95, color='red', linestyle='--', label='95% Variance Threshold')
plt.title('Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.legend()
plt.grid()
plt.show()
```

```python
pca_optimal = PCA(n_components=9)
pca_transformed = pca_optimal.fit_transform(X_scaled)
```

# 3. Clustering with Original Data

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Fit K-means
kmeans = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans.fit_predict(X_scaled)

# Evaluate clustering
silhouette = silhouette_score(X_scaled, clusters_original)
print(f'Silhouette Score (Original Data): {silhouette}')
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters_original, cmap='viridis', s=50)
plt.title('K-Means Clustering (Original Data)')
plt.show()
```

# 4. Clustering with PCA Data

```python
# Fit K-means on PCA-transformed data
kmeans = KMeans(n_clusters=3, random_state=42
clusters_pca = kmeans.fit_predict(pca_transformed)

# Evaluate clustering
silhouette_pca = silhouette_score(pca_transformed, clusters_pca)
print(f'Silhouette Score (PCA Data): {silhouette_pca}')
```

```python
plt.scatter(pca_transformed[:, 0], pca_transformed[:, 1], c=clusters_pca, cmap='viridis', s=50)
plt.title('K-Means Clustering (PCA Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show
```
