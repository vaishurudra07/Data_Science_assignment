```python
df=df.drop_duplicates()
```

```python
for i in df.isna().sum():
    print(round(i/len(df)*100,2),'%')
```

```python
 df.dropna(inplace=True)
```

```python
# Step 5: Detect and treat outliers using IQR method
numerical_columns = df.select_dtypes(include=[np.number]).columns

for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    print(f"Outliers in {col}: {outliers}")
```

# 2. Statistical Summary

```python
# Statistical summary of the dataset
summary = df.describe()

# Add median and IQR to the summary
summary.loc['median'] = df.median()
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
summary.loc['IQR'] = IQR

# Display the summary
print("\nStatistical Summary:")
print(summary)
```

# 3. Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Visualize distributions of numerical variables
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color='blue')
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Step 2: Boxplots for outlier detection
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# Step 3: Correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 4: Pair plot for all numerical variables
sns.pairplot(df[numerical_columns])
plt.show()
```

```python
# Step 5: Violin plot for a specific feature
plt.figure(figsize=(8, 6))
sns.violinplot(x='ASTV', y='LB', data=df)
plt.title("Violin Plot: ASTV vs LB")
plt.show()
```

# 4. Pattern Recognition and Insights

```python
# Step 1: Identify highly correlated features
high_correlation = correlation_matrix.abs().unstack().sort_values(ascending=False)
high_correlation = high_correlation[high_correlation < 1]  # Exclude self-correlation
print("\nTop Correlations Between Variables:")
print(high_correlation.head(10))
```

```python
# Step 2: Scatter plot for top correlated variables
plt.figure(figsize=(8, 6))
sns.scatterplot(x='LB', y='UC', data=df)
plt.title("Scatter Plot: LB vs UC")
plt.xlabel('Baseline FHR (LB)')
plt.ylabel('Uterine Contractions (UC)')
plt.show()
```
